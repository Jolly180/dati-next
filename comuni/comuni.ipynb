# Group By

```python
abitanti_per_regione = comuni.groupby('Sigla_Regione')['Abitanti'].sum()
abitanti_per_regione.head(3)
```

```python
abitanti_per_regione.plot(kind='bar', figsize=(12, 6), title='Abitanti per Regione')
```

```python
abitanti_per_regione.plot(kind='barh', figsize=(2, 8), title='Abitanti per Regione')
```

```python
abitanti_per_regione.plot(kind='pie', figsize=(12, 6), title='Abitanti per Regione')
```

```python
abitanti_per_regione.sort_values(ascending=True).plot(kind='area', figsize=(12, 6), title='Abitanti per Regione')
```

```python
gruppo_regione = comuni.groupby('Sigla_Regione')
somma_abitanti_regione = gruppo_regione['Abitanti'].sum()
num_province_regione   = gruppo_regione['Sigla_Provincia'].nunique()
num_comuni_regione     = gruppo_regione['Nome_Comune'].nunique()

# Crea una nuova tabella con una riga per ogni regione
regioni = pd.DataFrame({'somma_abitanti': somma_abitanti_regione,
                        'num_province': num_province_regione,
                        'num_comuni': num_comuni_regione})
regioni.head(10)
```
